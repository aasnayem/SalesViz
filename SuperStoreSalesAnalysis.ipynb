```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/aasnayemgazzalichowdhury/Desktop/SuperStoreOrders.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
# Convert date
df['order_date'] = pd.to_datetime(df['order_date'])
df['order_date']

df['ship_date'] = pd.to_datetime(df['ship_date'])
df['ship_date']
```

```python
# Create time feautures
df['month'] = df['order_date'].dt.month
df['month_name'] = df['order_date'].dt.month_name()
df['day_name'] = df['order_date'].dt.day_name()
```

```python
df
```

```python
# Convert sales column to numeric
df['sales'] = pd.to_numeric(df['sales'], errors='coerce')

# Check if any bad values became null
print(df['sales'].isnull().sum())

# Create customer spend
customer_spend = df.groupby('customer_name')['sales'].sum().reset_index()

customer_spend.rename(
    columns={'sales':'customer_total_sales'},
    inplace=True
)

df = df.merge(customer_spend,on='customer_name')

```

```python
df
```

```python
# Create customer segmentation

def segment_customer(x):
    if x > 5000:
        return 'High Value'
    elif x > 1000:
        return 'Medium Value'
    else:
        return 'Low Value'

df['customer_segment'] = df['customer_total_sales'].apply(segment_customer)

```

```python
df
```

```python
# Top products
print(
    df.groupby('product_name')['sales']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

# Sales by category
print(
    df.groupby('category')['sales']
    .sum()
)

# Sales by month
print(
    df.groupby('month_name')['sales']
    .sum()
)

# Customer segments count
print(
    df['customer_segment'].value_counts()
)

# Preview final dataset
print(df.head())
```

```python
df.groupby('category')['profit'].sum().sort_values(ascending=False)
df
```

```python
# Drop duplicate/corrupted columns
df = df.drop(
    columns=[
        'customer_total_sales_x',
        'customer_total_sales_y'
    ],
    errors='ignore'
)

# Check remaining columns
print(df.columns)
```

```python
# Export final cleaned dataset to Excel
df.to_csv("/Users/aasnayemgazzalichowdhury/Desktop/final_sales_dataset.csv", index=False)
```
